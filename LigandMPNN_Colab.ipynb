````python
import torch
import glob
import os
import numpy as np

# 1. 搜索 .pt 文件 (注意后缀名改了)
search_pattern = "./outputs/peg10_score/stats/*.pt"
files = glob.glob(search_pattern)

if files:
    pt_file = files[0]
    print(f"✅ 终于找到了！目标文件: {os.path.basename(pt_file)}")

    # 2. 使用 torch 加载数据
    data = torch.load(pt_file)

    # 3. 提取分数
    final_score = None

    # 优先提取 score_list
    if 'score_list' in data:
        # 获取第一个序列的得分
        raw_score = data['score_list'][0]
        # 如果是 Tensor 格式，转成数字
        if isinstance(raw_score, torch.Tensor):
            final_score = raw_score.item()
        else:
            final_score = raw_score

    # 备用：计算 log_probs
    elif 'log_probs' in data:
        lp = data['log_probs']
        if isinstance(lp, torch.Tensor):
            lp = lp.cpu().numpy()
        final_score = -np.mean(lp)

    # 4. 显示结果
    print("\n" + "="*40)
    print(f"  🧬 结合能力得分 (Score): {final_score:.4f}")
    print("="*40 + "\n")

    # 5. 结果解读
    print("【结果解读】")
    if final_score < -1.5:
        print("🌟 强结合 (Strong Binder)")
        print("   分数很低（负值大），说明蛋白序列非常匹配该 RNA 环境。")
````

````python
import numpy as np

def find_zinc_binding_residues(pdb_path, distance_cutoff=2.8):
    """自动寻找 PDB 文件中与 ZN 原子距离小于 cutoff 的氨基酸"""
    zinc_coords = []
    residues = {}

    with open(pdb_path, 'r') as f:
        for line in f:
            if line.startswith("HETATM") and "ZN" in line:
                # 提取锌离子坐标
                x = float(line[30:38])
                y = float(line[38:46])
                z = float(line[46:54])
                zinc_coords.append(np.array([x, y, z]))

            elif line.startswith("ATOM"):
                # 提取蛋白原子坐标
                chain_id = line[21]
                res_seq = line[22:26].strip()
                res_name = line[17:20].strip()
                atom_name = line[12:16].strip()

                # 只看侧链关键原子 (SG for Cys, ND1/NE2 for His)
                if res_name == "CYS" and "SG" in atom_name:
                    coords = np.array([float(line[30:38]), float(line[38:46]), float(line[46:54])])
                    key = f"{chain_id}{res_seq}" # 格式如 A23
                    residues[key] = {'name': res_name, 'coords': coords}
                elif res_name == "HIS" and ("ND1" in atom_name or "NE2" in atom_name):
                    coords = np.array([float(line[30:38]), float(line[38:46]), float(line[46:54])])
                    key = f"{chain_id}{res_seq}"
                    residues[key] = {'name': res_name, 'coords': coords}

    # 计算距离
    binding_sites = []
    if not zinc_coords:
        print("❌ 警告：PDB 文件里没找到 'ZN' (锌离子)！")
        return ""

    for zn in zinc_coords:
        for res_id, info in residues.items():
            dist = np.linalg.norm(zn - info['coords'])
            if dist < distance_cutoff:
                binding_sites.append(res_id)
                print(f"✅ 发现锌结合位点: {res_id} ({info['name']}) - 距离: {dist:.2f} Å")

    return " ".join(list(set(binding_sites))) # 去重并用空格连接

# 运行检测
pdb_file = "./inputs/peg10-RNA.pdb"
fixed_residues_str = find_zinc_binding_residues(pdb_file)

if fixed_residues_str:
    print("-" * 30)
    print(f"🔐 即将锁定的关键位点: {fixed_residues_str}")
    print("-" * 30)
else:
    print("⚠️ 没自动检测到。请手动确认锌离子是否存在，或手动指定位点。")
````

````python
# 设置输出文件夹
output_design = "./outputs/peg10_optimized"

# 1. 构造运行命令
# 注意：我们把 fixed_residues_str 传进去
cmd = f"""
python run.py \
    --model_type "ligand_mpnn" \
    --pdb_path "./inputs/peg10-RNA.pdb" \
    --out_folder "{output_design}" \
    --save_stats 1 \
    --ligand_mpnn_use_atom_context 1 \
    --fixed_residues "{fixed_residues_str}" \
    --batch_size 10 \
    --number_of_batches 1 \
    --temperature 0.1 \
    --verbose 0
"""

print(f"🚀 开始设计... (锁定位点: {fixed_residues_str})")
print("AI 正在尝试修复原子冲突并增强结合力...")
get_ipython().system(cmd)

# 2. 结果展示：看看 AI 建议突变哪里
import glob
import torch

# 找到生成的 .pt 文件
pt_files = glob.glob(f"{output_design}/stats/*.pt")

if pt_files:
    data = torch.load(pt_files[0])

    # 获取原生序列和设计序列
    # LigandMPNN 的 output 通常 native 是最后一个，或者在 seqs 文件夹里
    # 这里我们直接读生成的 fasta 更直观
    fa_files = glob.glob(f"{output_design}/seqs/*.fa")
    if fa_files:
        with open(fa_files[0], 'r') as f:
            lines = [l.strip() for l in f.readlines() if l.strip()]

        print("\n" + "="*60)
        print(f"{'排名':<5} | {'得分 (越低越好)':<15} | {'序列 & 突变建议'}")
        print("-" * 60)

        # 提取数据 (FASTA 格式: >sample=1, score=xyz ...)
        parsed = []
        for i in range(0, len(lines), 2):
            header = lines[i]
            seq = lines[i+1]
            # 提取分数
            try:
                score = float(header.split("score=")[1].split(",")[0])
            except:
                score = 99.9
            parsed.append((score, seq, header))

        # 排序
        parsed.sort(key=lambda x: x[0])

        # 原生序列通常作为对比（假设第一条是 sample 1，如果分数很高就是原生可能有问题）
        # 这里直接列出最好的前 5 个
        for idx, (score, seq, header) in enumerate(parsed[:5]):
            print(f"Top {idx+1:<2} | {score:<15.4f} | {seq}")

        print("="*60)
        print("💡 建议：")
        print("1. 找那些得分变成负数（如 -1.5）的序列。")
        print("2. 对比它和你的原序列哪里不一样，那就是你需要做点突变的位置！")
````
